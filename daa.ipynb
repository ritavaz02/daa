Agora vamos analizar a feature 'Injeção na rede'

```python
energia_2021['Injeção na rede (kWh)'].describe()
```

Injeção na rede pode ter 4 valores possíveis: Low(1), Medium(2), High(3), Very High(4)

```python
replace_map = {'injecao': {'None':0, 'Low':1, 'Medium':2, 'High':3, 'Very High':4}}
labels = energia_2021['Injeção na rede (kWh)'].astype('category').cat.categories.tolist()
replace_map_comp = {'injecao' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}
print(replace_map_comp)
```

```python
energia_2021.replace(replace_map_comp, inplace=True)
energia_2021.head()
```

```python
energia_2021['Injeção na rede (kWh)'].describe()
```

```python
print(energia_2021['Injeção na rede (kWh)'].dtypes)
```

```python
energia_2022.replace(replace_map_comp, inplace=True)
energia_2022.head()
print(energia_2022['Injeção na rede (kWh)'].dtypes)
```
